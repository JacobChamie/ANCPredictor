```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

```

```python
#Data Points, x = day, y = anc
x = [1, 2, 3, 4, 5, 6, 7]
y = [0, 0, 0, 10, 40, 90, 200]
```

```python
polynomialFit = numpy.poly1d(numpy.polyfit(x,y,3))
lineFit = numpy.linspace(1, x[-1], y[-1])
```

```python
plt.scatter(x, y)
plt.plot(lineFit, polynomialFit(lineFit))
plt.show()
```

```python
#R2 Score:
print(r2_score(y, polynomialFit(x)))
```

```python
#Predict future values:
day10 = polynomialFit(10)
print(day10)
```
